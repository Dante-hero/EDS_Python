**strip() function in Python:**

The strip() function in Python is a
built-in string method that is used to remove leading and trailing characters (whitespace or specified characters) from a string.It returns a new string with the desired characters stripped, while leaving the original string unchanged.


```python
#The general syntax of the strip() function is as follows:
string.strip([characters])
```

```python
# Example 1: Removing Whitespace
text = "   Hello, World!   "
stripped_text = text.strip()
print(stripped_text)
```

```python
# Example 2: Removing Specific Characters
text = "***Welcome to Python***"
stripped_text = text.strip("*")
print(stripped_text)
```

```python
# Example 3:
text = "***Hello***World***"
stripped_text = text.strip("*")
print(stripped_text)
#Options:
#A)Hello***World
#B)HelloWorld
```
